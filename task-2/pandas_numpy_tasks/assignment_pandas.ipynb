```python
import pandas as pd
```

```python
# загружаем датасет
data = pd.read_csv("./data/adult.data.csv")
data.head(5)
```

**1. Посчитайте, сколько мужчин и женщин (признак *sex*) представлено в этом датасете**

```python
female_count = data['sex'].value_counts()['Female']
male_count = data['sex'].value_counts()['Male']
print(f'Female count: {female_count}, Male count: {male_count}')
```

**2. Каков средний возраст мужчин (признак *age*) по всему датасету?**

```python
avg_male_age = data[data.sex == 'Male']['age'].mean()
print(f'Average male age: {avg_male_age}')
```

**3. Какова доля граждан Соединенных Штатов (признак *native-country*)?**

```python
us_citizens_part = data['native-country'].value_counts()['United-States'] / len(data)
print(f'US citizens take {round(us_citizens_part * 100)}% of the data')
```

**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

```python
more_50k = data[data.salary == '>50K']
avg_age = more_50k['age'].mean()
std_dev = more_50k['age'].std()
print(f'Average age: {avg_age}, standard deviation: {std_dev}')
```

**6. Правда ли, что люди, которые получают больше 50k, имеют минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
higher_education = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
more_50k_total_count = len(more_50k['education'])
more_50k_higher_ed_len = len(more_50k[more_50k.education.isin(higher_education)])
higher_ed_percentage = round(more_50k_higher_ed_len / more_50k_total_count * 100)
print(f'People with the higher education take {higher_ed_percentage}% of population making more than 50k a year')
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Asian-Pac-Islander*.**

```python
grouped_by_race = data.groupby(['race', 'sex'])
age_stats = grouped_by_race['age'].describe()
print('Age statistics: \n', age_stats)
max_age = int(data[(data.race == 'Asian-Pac-Islander') & (data.sex == 'Male')].groupby('race')['age'].describe()['max'].iloc[0])
print(f'\nMax age of male Asian-Pac-Islander - {max_age}')
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
married_values = ['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']
married  = data[(data['marital-status'].isin(married_values)) & (data['sex'] == 'Male')]
unmarried = data[(~data['marital-status'].isin(married_values)) & (data['sex'] == 'Male')]
married_more50k_part = married['salary'].value_counts()['>50K'] / len(married)
unmarried_more50k_part = unmarried['salary'].value_counts()['>50K'] / len(unmarried)
print(f'The part of married men making > 50K: {round(married_more50k_part * 100)}%')
print(f'The part of married men making > 50K: {round(unmarried_more50k_part * 100)}%')
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_working_hours = data['hours-per-week'].max()
max_workers = data[data['hours-per-week'] == max_working_hours]
workers_count = len(max_workers)

percents_of_good_paid = round(max_workers['salary'].value_counts()['>50K'] / workers_count * 100)

print(f'Max working hours - {max_working_hours}hr per week')
print(f'Count of people working maximum amount of time - {workers_count}')
print(f'Percentage of well-compensated - {percents_of_good_paid}%')
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
low_salary = '<=50K'
good_salary = '>50K'

undercompensated = data[data.salary == low_salary].groupby('native-country')
undercomp_avg_working_hours = undercompensated['hours-per-week'].mean()
print(f'Average working hours of undercompensated workers per country: \n{undercomp_avg_working_hours}')

well_compensated = data[data.salary == good_salary].groupby('native-country')
wellcomp_avg_working_hours = well_compensated['hours-per-week'].mean()
print(f'\nAverage working hours of well compensated workers per country: \n{wellcomp_avg_working_hours}')
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
data["AgeGroup"] = data['age'].case_when([
    (data['age'].isin(range(16, 35)), 'young'),
    (data['age'].isin(range(35, 70)), 'adult'),
    (data['age'].isin(range(70, 100)), 'retiree'),
])

data.head(20)
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
groups_count_more50k = data[data['salary'] == '>50K'].groupby(['AgeGroup', 'salary']).size()
print(f'Count of people making >50K a year for each age group: \n{groups_count_more50k}\n')

groups_count_total = data.groupby(['AgeGroup']).size()
percentage = groups_count_more50k / groups_count_total
name = percentage.idxmax()[0]
print(f'The group where the percentage of well-paid workers is the hihgest: {name}')
print(f'Percentage table: \n{percentage}')
```
